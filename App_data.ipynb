```python
# 8) Convert the column "Reviews'' to numeric data type and check the presence of outliers in the column and handle the outliers using a transformation approach.(Hint: Use log transformation)
app_data["Reviews"]=pd.to_numeric(app_data['Reviews'],errors='coerce') #changing reviews columns as object to numeric
```

```python
app_data['Reviews_log'] = np.log1p(app_data['Reviews']) #using log transform we are handling the ouliers in the columns
```

```python
app_data.boxplot(column='Reviews_log')
```

```python
# 9) The column 'Size' contains alphanumeric values, treat the non numeric data and convert the column into suitable data type. 
# (hint: Replace M with 1 million and K with 1 thousand, and drop the entries where size='Varies with device')
app_data['Size']
```

```python
app_data['Size']=app_data['Size'].replace('Varies with device',np.nan)
```

```python
app_data=app_data.dropna(subset=['Size'])
```

```python
app_data.isnull().sum()
```

```python
app_data['Size'].str[-1].unique()
```

```python
app_data['Size']=app_data['Size'].replace({'M':'*10**6','K':'*10**3','k':'*10**3'},regex=True)
```

```python
app_data['Size']=app_data['Size'][app_data['Size'].isnull()==False].map(pd.eval)
#eval() function is used to perform various operations on the dataframe columns, such as arithmetic operations, logical operations, and string operations
```

```python
app_data['Size']
```

```python
app_data['Size'].dtype
```

```python
# 10) Check the column 'Installs', treat the unwanted characters and convert the column into a suitable data type.
app_data['Installs'].unique()
```

```python
app_data['Installs']=app_data['Installs'].str.replace(',','')
```

```python
app_data['Installs']=app_data['Installs'].str.replace('+','')
```

```python
app_data['Installs']
```

```python
# 11) Check the column 'Price' , remove the unwanted characters and convert the column into a suitable data type.
app_data['Price'].unique()
```

```python
app_data['Price']=app_data['Price'].str.replace('$',"")
```

```python
app_data['Price'] = pd.to_numeric(app_data['Price'])
```
